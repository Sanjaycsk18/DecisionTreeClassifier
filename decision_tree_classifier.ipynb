### Classification Using Decision Tree Classifier

#### Importing Modules

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import train_test_split
```

#### Loading Dataset

```python
iris_df = pd.read_csv('./datasets/Iris.csv',index_col='Id')
iris_df.head()
```

#### Visualizing The Data

```python
sns.pairplot(iris_df, hue = 'Species' )
plt.show()
```

```python
x = iris_df.drop('Species',axis=1).values
print('Shape of X :', x.shape)
print('Total samples in the data : ',x.shape[0])
 
y = iris_df['Species']

train_x,test_x,train_y,test_y = train_test_split(x,y,test_size = 0.25)
print("No.of training samples : ",train_x.shape[0])
print("No.of testing samples : ",test_x.shape[0])


```

#### Implementing Decision Tree Classifier

```python
model = DecisionTreeClassifier()
model.fit(train_x,train_y)

print("Test Score : ",model.score(test_x,test_y))
```

#### Visualizing The Decision Tree

```python
features = ['Sepal Length(cm)','Sepal Width(cm)','Pedal Length(cm)','Pedal Width(cm)']
classes = ['Iris-setosa','Iris-versicolor','Iris-virginica']

plt.figure(figsize=(4,4),dpi = 800)
plot_tree(model,feature_names = features, class_names = classes,filled = True)
plt.show()
```
